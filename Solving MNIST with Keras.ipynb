```python
import pandas as pd
data=pd.read_csv('MNIST/train.csv', sep=',')
data_test=pd.read_csv('MNIST/test.csv',sep=',')
data.head
```

# CONSTRUIMOS entrenamiento y test

```python
import numpy as np
#construimos entrenamiento
y_train=data['label']
cols_pixel_train=data.drop(columns=['label']).to_numpy()
X_train=np.reshape(cols_pixel_train,(42000,28,28,1))



    
#construimos test
cols_pixel_test=data_test.to_numpy()
#no hay clase en test, se evalua online
X_test=np.reshape(cols_pixel_test,(28000,28,28,1))

    
```

```python
#primero con knn y regresion logistica

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=500).fit(cols_pixel_train, y_train)
```

```python
logreg.score(cols_pixel_train, y_train)
```

```python
file = open('MNIST/submission.csv', 'w')
head="ImageId,Label\n"
file.write(head)
for i in range(1,len(data_test)+1):
    clase=logreg.predict([cols_pixel_test[i-1]])
    file.write(str(i)+","+str(clase[0])+"\n")
    
file.close()
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Conv2D,MaxPooling2D,Flatten 



model = Sequential()

model.add(Conv2D(32,(3,3),padding='same',input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(Conv2D(32,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))



model.add(Conv2D(64,(3,3),padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator

gen=ImageDataGenerator(
    featurewise_center=False, samplewise_center=False,
    featurewise_std_normalization=False, samplewise_std_normalization=False,
    zca_whitening=False, zca_epsilon=1e-06, rotation_range=10, width_shift_range=0.1,
    height_shift_range=0.1, brightness_range=None, shear_range=0.0, zoom_range=0.0,
    channel_shift_range=0.0, fill_mode='nearest', cval=0.0, horizontal_flip=False,
    vertical_flip=False, rescale=None, preprocessing_function=None,
    data_format=None, validation_split=0.0, dtype=None
)
```

```python
gen.fit(X_train)
```

```python
model.fit(gen.flow(X_train,y_train),epochs=110,batch_size=32)
```

```python
predictions = model.predict(X_test)
```

```python
file = open('MNIST/submissionPRO.csv', 'w')
head="ImageId,Label\n"
file.write(head)
i=1
for prediction in predictions:
    file.write(str(i)+","+str(prediction.argmax())+"\n")
    i+=1
file.close()
```
